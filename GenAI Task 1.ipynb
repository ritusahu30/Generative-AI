```python
import yt_dlp
import requests
import json
from moviepy.editor import VideoFileClip
```

```python
# To download video from Youtube
url = "https://www.youtube.com/watch?v=2p5OHDxR2l8"

ydl_opts = {
    'format': 'best',  # Download the best quality available
    'outtmpl': '%(title)s.%(ext)s',  # Save the file with the video's title as the filename
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])

print("Video downloaded successfully!")
```

```python
# Load the video file
video_file_path = 'Introduction to Generative AI.mp4'  # Replace with your video file path
video = VideoFileClip(video_file_path)

# Extract audio and save it as an MP3 file
audio_file_path = 'Introduction to Generative AI.mp3'  # Specify the desired audio file name
video.audio.write_audiofile(audio_file_path, codec='mp3')
```

```python
ELEVEN_LABS_API_KEY = 'sk_011e06d09efc76025331e3fdcc7909341d7c81fbac6c4d25'
ELEVEN_LABS_URL ='https://elevenlabs.io/app/speech-synthesis/text-to-speech'

DEEPGRAM_API_KEY = 'f5203d291f2b7663d76855495b6dde2f85aab9c4'
DEEPGRAM_URL = 'https://api.deepgram.com/v1/listen'

audio_file_path = 'Introduction to Generative AI.mp3' 
```

```python
# Step 4: Set the headers for the API request (authorization with your API key)
headers = {
    'Authorization': f'Token {DEEPGRAM_API_KEY}',
    'Content-Type': 'application/mp3'
}
headers
```

```python
# Step 5: Send the audio file to Deepgram API for transcription
with open(audio_file_path, 'rb') as audio:
    response = requests.post(DEEPGRAM_URL, headers=headers, data=audio)

print(response)
```

```python
# Step 6: Handle the response
if response.status_code == 200:
    print("Request was successful.")
    transcription_response = response.json()  # Parse the JSON response
    print("Full Response:", transcription_response)
    
    # Step 7: Extract the transcription text
    try:
        transcription = transcription_response['results']['channels'][0]['alternatives'][0]['transcript']
        print("Transcription:", transcription)
    except KeyError:
        print("Error: Could not find transcription in response. Check the response structure.")
else:
    print(f"Request failed with status code {response.status_code}: {response.text}")
```

```python
# ElevenLab: The transcribed text from Deepgram
transcribed_text = transcription
transcribed_text
```

```python
# Step 4: Set the headers for Eleven Labs API request (authorization with your API key)
headers = {
    'Content-Type': 'application/json',
    'xi-api-key': ELEVEN_LABS_API_KEY  # Set your API key here
}
headers
```

```python
# Step 5: Create the payload with transcribed text and voice settings
payload = {
    'text': transcribed_text,  # The text to convert to speech
    'voice_id': 'default',  # Replace with the desired voice ID from Eleven Labs
    'model_id': 'eleven_monolingual_v1',  # Specific Eleven Labs model
    'settings': {
        'stability': 0.75,
        'similarity_boost': 0.75
    }
}
payload
```

```python
# Step 6: Send the POST request to Eleven Labs to convert text to speech
response = requests.post(ELEVEN_LABS_URL, headers=headers, data=json.dumps(payload))
response
```

```python
# Step 7: Handle the response
if response.status_code == 200:
    # Save the audio file
    audio_file_path = 'output_audio.mp3'  # Path to save the output audio file
    with open(audio_file_path, 'wb') as f:
        f.write(response.content)
    print(f"Audio saved as {audio_file_path}")
else:
    print(f"Error {response.status_code}: {response.text}")
```
